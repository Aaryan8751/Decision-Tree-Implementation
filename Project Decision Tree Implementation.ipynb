```python
from sklearn import datasets
import pandas as pd
import numpy as np
```

```python
iris = datasets.load_iris()
```

```python
df = pd.DataFrame(iris.data)
df.columns = ["sl", "sw", 'pl', 'pw']
```

```python
#Function to find label for a value
#if MIN_Value <=val < (m + Mean_Value) / 2 then it is assigned label a
#if (m + Mean_Value) <=val < Mean_Value then it is assigned label b
#if (Mean_Value) <=val < (Mean_Value + MAX_Value)/2 then it is assigned label c
#if (Mean_Value + MAX_Value)/2 <=val <= MAX_Value  then it is assigned label d

def label(val, *boundaries):
    if (val < boundaries[0]):
        return 'a'
    elif (val < boundaries[1]):
        return 'b'
    elif (val < boundaries[2]):
        return 'c'
    else:
        return 'd'

#Function to convert a continuous data into labelled data
#There are 4 lables  - a, b, c, d
def toLabel(df, old_feature_name):
    second = df[old_feature_name].mean()
    minimum = df[old_feature_name].min()
    first = (minimum + second)/2
    maximum = df[old_feature_name].max()
    third = (maximum + second)/2
    return df[old_feature_name].apply(label, args= (first, second, third))
```

```python
#Convert all columns to labelled data
df['sl_labeled'] = toLabel(df, 'sl')
df['sw_labeled'] = toLabel(df, 'sw')
df['pl_labeled'] = toLabel(df, 'pl')
df['pw_labeled'] = toLabel(df, 'pw')
df
```

```python
def entropy(X_train_column_data,Y_train,classe):
    m1 = Y_train[np.where(X_train_column_data==classe)]
    total = len(m1)
    entropy = 0
    values,counts = np.unique(m1,return_counts=True)
    p = counts/total
    log_p = np.log(p)
    p_log_p = -p*log_p
    return p_log_p.sum()

def gain(X_train,Y_train,f,d):
    d_classes = {}
    unique_classes_in_f = set(X_train[:,d[f]])

    values,counts = np.unique(X_train[:,d[f]],return_counts=True)
    for i in range(len(values)):
        d_classes[values[i]] = d_classes.get(values[i],0)+counts[i] 
    info_f = 0
    for classe in unique_classes_in_f:
        info_f += (d_classes[classe]/len(X_train)) * entropy(X_train[:,d[f]],Y_train,classe)
    values1,counts1 = np.unique(Y_train,return_counts=True)
    total = len(Y_train)
    p = counts1/total
    log_p = np.log(p)
    p_log_p = -p*log_p
    info_original = p_log_p.sum()
    information_gain = info_original - info_f
    return information_gain,info_original
        
def d_T(X_train,Y_train,features,d,level):
    
    if(len(set(Y_train))==1):
        print("Level",level)
        x=0
        y=0
        z=0
        if (Y_train[0]==0):
            x=len(Y_train)
        elif (Y_train[0]==1):
            y=len(Y_train)
        else:
            z=len(Y_train)
        print("Count of 0",x)
        print("Count of 1",y)
        print("Count of 2",z)
        print("Current Entropy  is =  0.0")
        print("Reached Leaf Node")
        return
    if len(features) == 0:
        print("Level",level)
        print("Features Ended")
        values12,counts12 = np.unique(Y_train,return_counts=True)
        d_counts = {0:0,1:0,2:0}
        for i in range(len(values12)):
            d_counts[values12[i]] = counts12[i]
        print("Count of 0",d_counts[0])
        print("Count of 1",d_counts[1])
        print("Count of 2",d_counts[2])
        
        return
    
    max_gain = -99999
    feature_split = 0
    entropy = 0
    print("Level",level)
    print("Count of 0",len(Y_train[np.where(Y_train==0)]))
    print("Count of 1",len(Y_train[np.where(Y_train==1)]))
    print("Count of 2",len(Y_train[np.where(Y_train==2)]))
    for f in features:
        info_gain ,entropy_level = gain(X_train,Y_train,f,d)
        if (max_gain<info_gain):
            max_gain = info_gain
            entropy = entropy_level
            feature_split = f
    
    print("Current Entropy",entropy)
    print("Splitting on feature {} with information gain {}".format(feature_split,max_gain))
    
    li = []
    for i in features:
        if i!=feature_split:
            li.append(i)
    li=np.array(li)
    unique_class_in_feature_split = set(X_train[:,d[feature_split]])
    for class1 in unique_class_in_feature_split:
        X_train_classe = X_train[np.where(X_train[:,d[feature_split]]==class1)]
        Y_train_classe = Y_train[np.where(X_train[:,d[feature_split]]==class1)]
        print("----------------------------------")
        X1 = d_T(X_train_classe,Y_train_classe,li,d,level+1)
        
```
